```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#plt.style.use()
# !pip install plotly
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

## Plot a scatter plot between total_bill and tip

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data)
```

# The Hue Parameter
* Hue use for categorical data
* style also use for Categorical data

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',kind='scatter',data=data)  # smoker and non smoker tip
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex',kind='scatter',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',kind='scatter',data=data) 
```

# The size parameter
* size parameter is use for Numerical Data (continuous and Discreate data)

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(15,200),kind='scatter',data=data) 
```

# Using the Scatterplot function


```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
```

## Plotting with categorical Data
* Categorical Scatterplots


* Stripplot
* Swarmplot
 

```python
plt.style.use('fivethirtyeight')
data.head()
```

```python
# x-axis------> categorical data
# y-axis------> Numerical Data

sns.catplot(x='day',y='tip',kind='strip',jitter=1,data=data) # jitter work like combine the data  
```

```python
sns.catplot(x='day',y='tip',kind='swarm',data=data)
```

## Boxplots
* # Distribution Plot

* use for numerical data 
* use for compare to two categorical data


```python
sns.boxplot(data['tip']) # Univariate
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data=data)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='box',data=data)
```

## Violin Plots
* it show the numerical data and can be considered a combination of the box plot with kernal(probability) density plot
* gives box plot details and entire details of the data


```python
sns.violinplot(data['total_bill']) 
```

```python
sns.catplot(x='day',y='total_bill',kind='violin',data=data) # 
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',data=data)
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='violin',split=True,data=data)
```

## Barplot and Countplot

* gives details about distribution Data 
* gives details about central tendency

```python
# find the mean of smoker tips and non smoker tips

sns.catplot(x='smoker',y='tip',kind='bar',data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',data=data)
```

```python
sns.catplot(x='smoker',y='total_bill',hue='sex',kind='bar',data=data)
```
